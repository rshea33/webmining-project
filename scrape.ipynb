```python
# click on each movie to get the movie details
url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
driver = webdriver.Safari()
driver.get(url)
soup = BeautifulSoup(driver.page_source, 'html.parser')


master_frame = pd.DataFrame()
titles = []
years = []
lengths = []
ratings = []
genres = []
directors = []

links = soup.find_all('a', href=True)
for link in links:
    if link['href'].startswith('/title/tt'):
        driver.get('https://www.imdb.com' + link['href'])
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        data = soup.find('title').text
        
        title = data.split('(')[0].strip()

        year = data.split('(')[1].split(')')[0].strip()
        year = int(year)

        li = soup.find_all('li', class_='ipc-inline-list__item')
        
        length = li[5].text
        

        director = li[6].text




        print(title, year, length, director)
        

        # length = soup.find('time').text
        # length = int(length.split(' ')[0])
        

        # genre: <span class="ipc-chip__text">Drama</span>
        # genre = soup.find('span', class_='ipc-chip__text').text
        
        # genre = soup.find('div', class_='subtext').text
        # director = soup.find('div', class_='credit_summary_item').text
        titles.append(title)
        years.append(year)
        lengths.append(length)
        # ratings.append(rating)
        # genres.append(genre)
        directors.append(director)
        time.sleep(4)

driver.close()

master_frame = pd.DataFrame()
master_frame['title'] = titles
master_frame['year'] = years
master_frame['length'] = lengths
# master_frame['rating'] = ratings
# master_frame['genre'] = genres
master_frame['director'] = directors
master_frame
```

```python
def convert_to_mins(length):
    # format of "xh ym"
    if len(length.split()) == 2:
        hours = int(length.split()[0].replace('h', ''))
        mins = int(length.split()[1].replace('m', ''))
        return hours * 60 + mins
    # format of "ym" or "xh"
    else:
        if 'h' in length:
            hours = int(length.split()[0].replace('h', ''))
            return hours * 60
        else:
            mins = int(length.split()[0].replace('m', ''))
            return mins
```
