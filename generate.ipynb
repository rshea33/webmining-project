```python
from PyMovieDb import IMDB
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('movies.csv')
titles = movies['movie_title']
titles
```

```python
import json
imdb = IMDB()

title = titles[0]

res = imdb.get_by_name(title) 
res = json.loads(res)
res
```

```python
res['review']
```

```python
def return_data(title):
    res = imdb.get_by_name(title) 
    res = json.loads(res)
    name = res['name']
    description = res['description']
    review = res['review']
    review_text = review['reviewBody']
    review_rating = review['reviewRating']['ratingValue']

    genre = res['genre']
    keywords = res['keywords']
    keywords = keywords.split(',')
    datePublished = res['datePublished']
    director = res['director']
    director = [d['name'] for d in director]

    actor = res['actor']
    actor = [a['name'] for a in actor]
    creator = res['creator']
    creator = [c['name'] for c in creator]

    return [name, description, review_text, review_rating, genre, keywords, datePublished, director, actor, creator]
```

```python
return_data(title)
```

```python
result = pd.DataFrame(columns=['name', 'description', 'review_text', 'review_rating', 'genre', 'keywords', 'datePublished', 'director', 'actor', 'creator'])
count = 0
for title in titles:
    try:
        result.loc[len(result)] = return_data(title)
        print(title, 'done')
        count += 1
        print(count, 'out of', len(titles), 'done')
    except:
        print(title, 'failed')
        count += 1
        print(count, 'out of', len(titles), 'done')
```
