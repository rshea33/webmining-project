```python
def return_data(title):
    res = imdb.get_by_name(title) 
    res = json.loads(res)
    name = res['name']
    description = res['description']
    review = res['review']
    review_text = review['reviewBody']
    review_rating = review['reviewRating']['ratingValue']

    genre = res['genre']
    keywords = res['keywords']
    keywords = keywords.split(',')
    datePublished = res['datePublished']
    director = res['director']
    director = [d['name'] for d in director]

    actor = res['actor']
    actor = [a['name'] for a in actor]
    creator = res['creator']
    creator = [c['name'] for c in creator]

    return [name, description, review_text, review_rating, genre, keywords, datePublished, director, actor, creator]
```

```python
return_data(title)
```

```python
result = pd.DataFrame(columns=['name', 'description', 'review_text', 'review_rating', 'genre', 'keywords', 'datePublished', 'director', 'actor', 'creator'])
count = 0
for title in titles:
    try:
        result.loc[len(result)] = return_data(title)
        print(title, 'done')
        count += 1
        print(count, 'out of', len(titles), 'done')
    except:
        print(title, 'failed')
        count += 1
        print(count, 'out of', len(titles), 'done')
```

```python
# begin to process data for modeling
df = pd.read_csv('updated_movies_no_na.csv')
df
```

```python
# remove all punctuation
import string
fn = lambda x: x.translate(str.maketrans('', '', string.punctuation)).lower()

df['description'] = df['description'].apply(fn)
df['review_text'] = df['review_text'].apply(fn)
df['genre'] = df['genre'].apply(fn)
df['keywords'] = df['keywords'].apply(fn)
df['director'] = df['director'].apply(fn)
df['actor'] = df['actor'].apply(fn)
df['creator'] = df['creator'].apply(fn)
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

# create tfidf vectorizer
tfidf = TfidfVectorizer(stop_words='english')
ct = CountVectorizer(stop_words='english')

# fit and transform on description
tfidf_des = pd.DataFrame(tfidf.fit_transform(df['description']).toarray())
tfidf_des.index = df['name']
tfidf_rev = pd.DataFrame(tfidf.fit_transform(df['review_text']).toarray())
tfidf_rev.index = df['name']
ct_genre = pd.DataFrame(ct.fit_transform(df['genre']).toarray())
ct_genre.index = df['name']
tfidf_keywords = pd.DataFrame(tfidf.fit_transform(df['keywords']).toarray())
tfidf_keywords.index = df['name']
ct_director = pd.DataFrame(ct.fit_transform(df['director']).toarray())
ct_director.index = df['name']
ct_actor = pd.DataFrame(ct.fit_transform(df['actor']).toarray())
ct_actor.index = df['name']
ct_creator = pd.DataFrame(ct.fit_transform(df['creator']).toarray())
ct_creator.index = df['name']
```

```python
# clustering algorithm
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import random

def return_similar_movies(movie_name, n_movies=5, data=tfidf_des):
    nn = NearestNeighbors(n_neighbors=n_movies, algorithm='ball_tree')
    nn.fit(data)
    # get the index of the movie
    idx = data.index.get_loc(movie_name)
    # get the 5 most similar movies
    similar_movies = nn.kneighbors(data.iloc[idx, :].values.reshape(1, -1), return_distance=False)
    # get the names of the movies
    similar_movies = [data.index[i] for i in similar_movies[0]]
    return similar_movies
```

```python
random.seed(1)
def test_similarity(data):
    for i in range(5):
        movie = random.choice(data.index)
        print('Movie selected:', movie)
        print('Similar movies:')
        print(return_similar_movies(movie, data=data))
        print()

test_similarity(tfidf_des)
```

```python
# add all dataframes together
master = pd.concat([tfidf_des, tfidf_rev, ct_genre, tfidf_keywords, ct_director, ct_actor, ct_creator], axis=1)
master.shape
```

```python
test_similarity(master)
```

```python
def test_kmeans(data, k=5):
    km = KMeans(n_clusters=k)
    km.fit(data)
    print('Inertia:', km.inertia_)
    print('Silhouette:', silhouette_score(data, km.labels_))
    print()
```
